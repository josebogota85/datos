EMPECEMOS A TRABAJAR CON LOS DATOS, CON PANDAS, VCAMOS A TRAER LOS DATOS DUPLICADOS

```python
leer_otro_Colombia['Year(s)'].unique() #Acá traemos una colección de datos con los datos únicos de una columna, por ejemplo si hay repetidos el los unifica
```

```python
#Supongamos que ahora queremos trabajar con los datos solo de cierto ranggoo, para eso usaremos algo como un filtro, veamoslo
leer_otro_Colombia['Year(s)']>=2022 #Aunque esperaba algo diferente lo que se puede ver es que a los datos que cumplen la condición les pone True y a los demas les pone False
```

```python
#También podemos seleccionar las columnas especificas de una linea, parece que nos va a mostrar ahora si lo que queriamos en lo anteriorr
leer_otro_Colombia_1=leer_otro_Colombia[leer_otro_Colombia['Year(s)']>=2003]#Ahora si esta haciendo el filtro, por que nos esta mostrando los registros filtrados por la condición que les dimos a las columnas, ojo que para hacer esto tuvimos que uardar el filtrto otra varable
print(leer_otro_Colombia_1)#Esto imprime los datos
```
