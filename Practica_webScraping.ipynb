```python
#Ahora usemos la información que extrajimos
import json #Primero importamos json para 
with open('apple.json') as json_file: #Ahora usamos open para abrir el archivo  y se le asigna a un objeto que se llama json_file, while sirve para que el archivo se cierre apenas se termine de leer
    apple_info = json.load(json_file)#Luego con json.load(json_file), se cargan los archivos en el objeto apple_info 
    # Print the type of data variable    
    #print("Type:", type(apple_info))
apple_info #Acá se muestra el arreglo que quedo dentro de apple_info
```

```python
apple_info['country'] #Acá traemos solo el dato que tenga la llave país country
#Este formato en que se devuelven los datos es DataFrame Pandas
```

```python
apple_share_price_data = apple.history(period="max") #Acá estamos usando el metodo .history para traer datos de la accion de apple desde el principio hasta hoy
apple_share_price_data #Este formato en que se devuelven los datos es DataFrame Pandas
```

```python
apple_share_price_data  = apple.history(period="1mo") #Acá por ejemplo traje los costos de la acción el último mes
apple_share_price_data #Este formato en que se devuelven los datos es DataFrame Pandas
```

```python
apple_share_price_data.head() #Al igual que pasa con los DataFrame de pandas cuando uso los metodos en esta tabla fucniona
```

```python
apple_share_price_data.reset_index(inplace=True) #Acá estamos reseteando el indice del DataFrame esto se hace para cuando se manipulan los datos se vuelva a poner el indice en cero
```

```python
apple_share_price_data.plot(x="Date", y="Open") #Así podemos graticar el precio de la acción vs la fecha
```
