```python
class rectangulo(object):#Creamos la clase rectangulo y los atributos de esta clase son: color, alto y ancho
    def __init__(self, color, alto, ancho):
        self.color=color;
        self.alto=alto;
        self.ancho=ancho;
    def mas_grande(self,ancho_1): # De esta manera hemos agregado un nuevo atributo que se sumará al ancho anterior, con el fin de agrandar el rectangulo
        self.ancho=self.ancho+ancho_1

rectangulo1=rectangulo("rojo",3,5) #Acá definimoms un objeto que es el primer rectangulo, ahora vamos a agragarle mas a su ancho
rectangulo1.ancho
rectangulo1.mas_grande(10)#De esta manera solo estamos cogiendo la función de la parte de abajo (mas_grande) y la estamos modificando para que el ancho aumente
rectangulo1.ancho
```
