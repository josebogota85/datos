```python
#Primero instalamos la biblioteca con pip install Pycoingecko
#Ahora lo que hacemos e importar la librería
from pycoingecko import CoinGeckoAPI # esta librería nos sirve para traer la información de la página CoinGeckoAPI
import pandas as pd #Acá estamos importando la librería de pandas
import matplotlib.pyplot as plt #Acá estamos importando la libredia de marplotlib
import plotly.graph_objs as go #Acá traemos el 
from plotly.offline import plot
import requests
cg = CoinGeckoAPI() #Creamos el objeto cliente #Por favor revisar como se crea un objeto


```

```python
bitcoin_data= cg.get_coin_market_chart_by_id(id='bitcoin', vs_currency='usd', days=30 ) #esta funcion nos ayuda a saber que datos vamos a solicitar
#Particularmente en la anterior función estamos obteniendo datos de bitctoin en dolares, para los ultimos 30 días
#print(datos_bitcoin) #Ojo que acá nos trae la info como un diccionario, lo cual nos ayuda a generare un dataframe lo recuerdan?
datos = pd.DataFrame(bitcoin_data['prices'], columns=['timestamp', 'price']) #Convertimos el diccionario en un dataframe
bitcoin_data['prices']
datos #ahora si tengo en datos la fecha y el precio, pero sucede que la fecha no se entiende mucho, así quye lo que vamos a hacer es darle un formato que se entienda
```

```python
#Vamos a usar la función de pandas .to_datetime, para darle formato a la fecha de la siguiente manera
datos['DATE'] = pd.to_datetime(datos['timestamp'], unit='ms') #Acá hizo una eliminación de dinamización, es decir agrupo por fecha pero con el dato minimo y maximo de cada día

datos
```

```python
#Ahora la ide es crear un grafico de velas, para eso tenemos que tener el maximo y el minimo diario, veamos
clandestick_data=datos.groupby(datos.DATE.dt.date).agg({'price':['min','max','first','last']})
clandestick_data
```

```python
fig = go.Figure(data=[go.Candlestick(x=clandestick_data.index,
                                      open=clandestick_data['price']['first'],
                                      high=clandestick_data['price']['max'],
                                      low=clandestick_data['price']['min'],
                                      close=clandestick_data['price']['last'])])#Esto es re confuso pero lo que hice en realidad fue como crear un objeto html, que muestra un gratico de velas, vamos a ver
fig.update_layout(xaxis_rangeslider_visible=False, xaxis_title='Date', yaxis_title='Price (USD)', title='Gráfica de Bit coin ultimos 30 días')

plot(fig, filename='bitcoin_candlestick_graph.html')

```
